```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

from reportlab.platypus import (
    SimpleDocTemplate,
    Paragraph,
    Spacer,
    Image,
    PageBreak
)
from reportlab.lib.styles import getSampleStyleSheet



file_path = input("Enter CSV file path: ")

try:
    df = pd.read_csv(file_path)

    print("\nDataset Loaded Successfully!\n")
    print(df.head())

except Exception as e:
    print("Error:", e)
    exit()



rows, cols = df.shape
missing_values = df.isnull().sum()

numeric_cols = df.select_dtypes(include="number").columns
cat_cols = df.select_dtypes(include="object").columns



hist_file = None

if len(numeric_cols) > 0:

    plt.figure(figsize=(6,4))
    df[numeric_cols[0]].hist(bins=20)

    plt.title(f"Histogram - {numeric_cols[0]}")
    plt.xlabel(numeric_cols[0])
    plt.ylabel("Frequency")

    hist_file = "histogram.png"

    plt.savefig(hist_file, bbox_inches="tight")
    plt.close()



scatter_file = None

if len(numeric_cols) >= 2:

    plt.figure(figsize=(6,4))

    plt.scatter(
        df[numeric_cols[0]],
        df[numeric_cols[1]]
    )

    plt.xlabel(numeric_cols[0])
    plt.ylabel(numeric_cols[1])

    plt.title(
        f"{numeric_cols[0]} vs {numeric_cols[1]}"
    )

    scatter_file = "scatter.png"

    plt.savefig(scatter_file, bbox_inches="tight")
    plt.close()



pie_file = None

if len(cat_cols) > 0:

    counts = df[cat_cols[0]].value_counts().head(5)

    plt.figure(figsize=(5,5))

    counts.plot(
        kind="pie",
        autopct="%1.1f%%"
    )

    plt.ylabel("")
    plt.title(f"Pie Chart - {cat_cols[0]}")

    pie_file = "pie.png"

    plt.savefig(pie_file, bbox_inches="tight")
    plt.close()



heatmap_file = None

if len(numeric_cols) > 1:

    plt.figure(figsize=(6,5))

    sns.heatmap(
        df[numeric_cols].corr(),
        annot=True,
        cmap="coolwarm"
    )

    plt.title("Correlation Heatmap")

    heatmap_file = "heatmap.png"

    plt.savefig(heatmap_file, bbox_inches="tight")
    plt.close()



print("\n" + "="*50)
print("AUTOMATIC DATASET ANALYSIS REPORT")
print("="*50)

print("\nDATASET INFORMATION")
print("-"*30)
print("Shape :", df.shape)
print("Rows :", rows)
print("Columns :", cols)

print("\nSAMPLE DATA")
print("-"*30)
print(df.head())

print("\nCOLUMN INFORMATION")
print("-"*30)

for col in df.columns:
    print(f"{col} : {df[col].dtype}")

print("\nMISSING VALUES")
print("-"*30)
print(missing_values)

print("\nSTATISTICAL SUMMARY")
print("-"*30)

for col in numeric_cols:

    print(f"\n{col}")
    print(f"Mean   : {df[col].mean():.2f}")
    print(f"Median : {df[col].median():.2f}")
    print(f"Min    : {df[col].min()}")
    print(f"Max    : {df[col].max()}")
    print(f"Std    : {df[col].std():.2f}")

print("\nIMPORTANT FINDINGS")
print("-"*30)

print("Total Records :", rows)

if len(numeric_cols) > 0:

    target_col = numeric_cols[-1]

    print(
        f"Average {target_col} :",
        round(df[target_col].mean(), 2)
    )

    print(
        f"Highest {target_col} :",
        df[target_col].max()
    )

    print(
        f"Lowest {target_col} :",
        df[target_col].min()
    )

if len(cat_cols) > 0:

    print(
        f"Most Common {cat_cols[0]} :",
        df[cat_cols[0]].mode()[0]
    )

print("\nGRAPHS GENERATED")
print("-"*30)

if hist_file:
    print("✓ Histogram")

if scatter_file:
    print("✓ Scatter Plot")

if pie_file:
    print("✓ Pie Chart")

if heatmap_file:
    print("✓ Correlation Heatmap")



pdf = SimpleDocTemplate("Report.pdf")
styles = getSampleStyleSheet()
content = []

content.append(
    Paragraph(
        "AUTOMATIC DATASET ANALYSIS REPORT",
        styles["Title"]
    )
)

content.append(Spacer(1,12))

content.append(
    Paragraph(
        f"Shape : {df.shape}",
        styles["Normal"]
    )
)

content.append(
    Paragraph(
        f"Rows : {rows}",
        styles["Normal"]
    )
)

content.append(
    Paragraph(
        f"Columns : {cols}",
        styles["Normal"]
    )
)

content.append(PageBreak())

content.append(
    Paragraph(
        "Missing Values",
        styles["Heading1"]
    )
)

for col, val in missing_values.items():

    content.append(
        Paragraph(
            f"{col} : {val}",
            styles["Normal"]
        )
    )

content.append(PageBreak())

content.append(
    Paragraph(
        "Statistical Summary",
        styles["Heading1"]
    )
)

for col in numeric_cols:

    content.append(
        Paragraph(
            f"<b>{col}</b>",
            styles["Heading2"]
        )
    )

    content.append(
        Paragraph(
            f"Mean : {df[col].mean():.2f}",
            styles["Normal"]
        )
    )

    content.append(
        Paragraph(
            f"Median : {df[col].median():.2f}",
            styles["Normal"]
        )
    )

    content.append(
        Paragraph(
            f"Minimum : {df[col].min()}",
            styles["Normal"]
        )
    )

    content.append(
        Paragraph(
            f"Maximum : {df[col].max()}",
            styles["Normal"]
        )
    )

    content.append(
        Paragraph(
            f"Std : {df[col].std():.2f}",
            styles["Normal"]
        )
    )

content.append(PageBreak())

content.append(
    Paragraph(
        "Graphs",
        styles["Heading1"]
    )
)

for img in [hist_file, scatter_file, pie_file, heatmap_file]:

    if img and os.path.exists(img):

        content.append(
            Image(
                img,
                width=400,
                height=250
            )
        )

        content.append(
            Spacer(1,10)
        )

pdf.build(content)

print("\nReport Generated Successfully!")
print("Saved as Report.pdf")
```
